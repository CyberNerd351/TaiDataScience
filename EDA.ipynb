```python
data = pandas.read_csv("datasets/supermarket_data.csv")
data
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

sns.countplot(x='gender', data=data, palette='pastel', width=0.5)
```

```python

sns.countplot(x='product', data=data, palette='winter', width=0.5)
plt.xticks(rotation=45)
```

```python
sns.barplot(x="gender",y="total",data=data,palette="summer")
```

```python
data['payment'].value_counts().plot.pie(autopct='%1.2f%%')
```

```python
import matplotlib.dates as mdates
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b')) # 'Jan', 'Feb', etc.
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
sns.lineplot(x='date', y='total', data=data)
```
