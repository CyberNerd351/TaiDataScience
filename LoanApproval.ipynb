```python
import pandas
data = pandas.read_csv("datasets/loan_approval.csv")
data
```

```python
data.describe()
```

```python
array = data.values
X = array [:,0:9]
Y = array [:,10]
```

```python
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.30, random_state=43)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)
print('Model Predicted', predictions)
print('Expected', Y_test)
```

```python
from sklearn.metrics import accuracy_score
print ('Accuracy', accuracy_score(Y_test, predictions))
```

```python
loan_approval = [[13, 2, 1, 158325, 34, 203385, 127949, 0, 85]]
outcome = model.predict(loan_approval)
print('Loan Approval', outcome)
```
